```python
import numpy as np 
import pandas as pd 
pd.set_option('display.max_columns', None)

data = pd.read_csv('/Users/ayonrab/Desktop/Econometrics Portfolio/LoanData/ujse_a_1434342_sm1547/SBAnational.csv')
data.head()
```

```python
#dropping whole row if MIS_Status = nan
print("Rows before drop: ", data.shape[0])
data = data.dropna(subset=['MIS_Status'])
print("Rows after drop: ", data.shape[0])
```

```python
#creating default variable (1 if MIS_Status = ChgOff, 0 if MIS_Status = P I F), along with some other variables
print(data['MIS_Status'].unique()) 
data['Default'] = np.where(data['MIS_Status']=='P I F', 0, 1)
data['New'] = np.where(data['NewExist']== 1, 0, 1)
data['RevLineCr'] = np.where(data['RevLineCr']== 'Y', 1, 0)
data['LowDoc'] = np.where(data['LowDoc']== 'Y', 1, 0)


#turning dollar amounts into floats
data[['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv']] = data[['DisbursementGross', 'BalanceGross', 'ChgOffPrinGr', 'GrAppv', 'SBA_Appv']].replace({"\$" : '', ',': ''}, regex = True)

#turning int columns into floats

data['DisbursementGross'] = [float(i) for i in data['DisbursementGross']]
data['BalanceGross'] = [float(i) for i in data['BalanceGross']]
data['ChgOffPrinGr'] = [float(i) for i in data['ChgOffPrinGr']]
data['GrAppv'] = [float(i) for i in data['GrAppv']]
data['SBA_Appv'] = [float(i) for i in data['SBA_Appv']]

#Remove this annoying year
data = data[data['ApprovalFY'] != '1976A']
data['ApprovalFY'] = [float(i) for i in data['ApprovalFY']]

data.head()
```

```python
#dropping unnecessary columns 
data_for_model = data.drop(['LoanNr_ChkDgt', 'Name', 'ApprovalDate', 'FranchiseCode', 'ChgOffDate', 'ChgOffPrinGr', 'MIS_Status', 'NewExist', 'DisbursementDate'], axis = 1)
print(data_for_model.shape)
```

```python
#Lets run a model for the state of Virginia and borrowers in VA
va_model = data_for_model[(data_for_model['State']=='VA')].copy()

#print(va_model.describe()) --> Saw that Max zip code was 99999, 24658 is the largest zipcode in VA
va_model = va_model[va_model['Zip'] < 24659]
print(va_model.shape)
va_model.describe()
```

```python
#looking at our VA Dataframe
va_model.head()
```

```python
#Dataframe is now clean and ready to be analyzed
#Lets get_dummies :) 
#for curiosity's sake, lets look at unique values first 
print("Top Loan Cities:", "\n", va_model['City'].value_counts().head())
print()
print("Top Loan Banks:", "\n",va_model['Bank'].value_counts().head())
print()
print("Top Loaned Industries:", "\n",va_model['NAICS'].value_counts().head())
print()
print("Top Loaned Years:", "\n",va_model['ApprovalFY'].value_counts().head())
```

```python
#use pandas.get_dummies()
va_model = pd.get_dummies(va_model, prefix = ['City','State', 'Bank', 'BankState'])
va_model.head()
```

```python
#Making a Random Forest Baseline
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn import metrics

x_cols = [col for col in va_model.columns if col !='Default']
x = va_model[x_cols]
y = va_model['Default']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .33, random_state = 0)

print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
#fitting baseline model
model = RandomForestClassifier()
model.fit(x_train, y_train)
```

```python
pred = model.predict(x_test)
accuracy = metrics.accuracy_score(y_test, pred)
print("Baseline accuracy is at: ", accuracy)
#pretty impressive baseline! Maybe we can improve it? 
```

```python
#Adding some relevant variables
va_model['Recession'] = np.where((va_model['ApprovalFY'] > 2007) & (va_model['ApprovalFY'] <= 2009), 1, 0)
va_model['RealEstate'] = np.where(va_model['Term'] >= 240, 1, 0)

#running the model again!
#but lets create a pipeline, so it will be faster in the future
def RandomForestPipeline(dataframe, y):
    x_cols = [col for col in dataframe.columns if col != y]
    x = dataframe[x_cols]
    y = dataframe[y]

    x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = .33, random_state = 0)
    
    model = RandomForestClassifier()
    model.fit(x_train, y_train)
    pred = model.predict(x_test)
    accuracy = metrics.accuracy_score(y_test, pred)
    
    return(pred, accuracy, x_train, y_train, x_test, y_test)

outcome = RandomForestPipeline(va_model, 'Default')
print("New Accuracy: ", outcome[1])
```

```python
#no change in accuracy! Maybe these new variables aren't helpful... Lets try hypertuning the parameters of the model 
from pprint import pprint 
print('Parameters currently in use:\n')
pprint(model.get_params())
```

```python
from sklearn.model_selection import RandomizedSearchCV# Number of trees in random forest

n_estimators = [int(x) for x in np.linspace(start = 100, stop = 2000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 110, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
random_search = RandomizedSearchCV(estimator = model, param_distributions = random_grid, n_iter = 10, cv = 3, verbose=2, random_state=42, n_jobs = -1)

x_train = outcome[2]
y_train = outcome[3]
random_search.fit(x_train, y_train)
random_search.best_params_
```

```python
#using new hyperparameters to predict outcome
x_test = outcome[4]
y_test = outcome[5]
pred = random_search.predict(x_test)
print(metrics.accuracy_score(y_test, pred))
```
